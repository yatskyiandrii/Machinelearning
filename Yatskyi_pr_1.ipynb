# Завдання 1 (Варіант 28)
 Створюємо одновимірний масив зі значеннями n! (факторіал) кожного натурального числа n

```python
import numpy as np
import math

# Кількість елементів у векторі
k = 10
factorials = np.array([math.factorial(i) for i in range(1, k+1)], dtype=object)

print("Вектор факторіалів n! для n = 1..10:\n", factorials)
```

# Завдання 2 (Варіант 28)

# Завдання 3 (Варіант 28)
 Масив 3x3 з випадковими числами від -1 до 1. Обчислити відсоток від’ємних.

```python
import numpy as np

A = np.random.uniform(-1, 1, (3,3))
neg_percent = np.sum(A < 0) / A.size * 100

print("Матриця A:\n", A)
print(f"\nВідсоток від’ємних елементів: {neg_percent:.2f}%")
```

# Завдання 8 (Варіант 28)
 Створюємо матрицю 3x3 з випадковими цілими числами 1–15 і знаходимо обернену матрицю.

```python
import numpy as np

A = np.random.randint(1, 16, (3,3)).astype(float)
det = np.linalg.det(A)

print("A:\n", A)
print("\nВизначник det(A) =", det)

if abs(det) < 1e-8:
    print("\nМатриця вироджена — оберненої не існує.")
else:
    A_inv = np.linalg.inv(A)
    print("\nОбернена матриця A^-1:\n", A_inv)
```

# Завдання 14 (Варіант 28)
 Створюємо матрицю 4x4 з випадковими цілими числами 1–100
 та знаходимо суму елементів побічної діагоналі.

```python
import numpy as np

n = 4
M = np.random.randint(1, 101, (n,n))
anti_diag_sum = sum(M[i, n-1-i] for i in range(n))

print("Матриця M:\n", M)
print("\nСума побічної діагоналі:", anti_diag_sum)
```

# Завдання 15 (Варіант 28)
Створюємо матрицю 3x3 з випадковими дійсними числами від -2 до 2.
 Знаходимо найменші за модулем елементи у кожному стовпці та формуємо вектор-стовпчик.

```python
import numpy as np

M = np.random.uniform(-2, 2, (3,3))
idx = np.argmin(np.abs(M), axis=0)
v = M[idx, range(M.shape[1])].reshape(-1,1)

print("Матриця M:\n", M)
print("\nІндекси мінімумів за модулем:", idx)
print("\nВектор-стовпчик мінімальних за модулем значень:\n", v)
```
